```python
import os
import json

folders = ["Clickbait_Auto", "Clickbait_Direct", "NonClickbait_Auto"]
categories = ["EC", "ET", "GB", "IS", "LC", "PO", "SO"]

# Function to extract newsTitle and newsContent from a file
def extract_values_from_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
        news_title = data['sourceDataInfo']['newsTitle']
        news_content = data['sourceDataInfo']['newsContent']
        return news_title, news_content

# Directory where your files are located
folder_path = './data/Part1'

# Lists to store news titles and news contents
data = [["title", "content", "label"]]

# Iterate through each file in the folder
for folder in folders:
    for category in categories:
        path = folder_path+"/"+folder+"/"+category
        for filename in os.listdir(path):
            file_path = os.path.join(path, filename)
            if os.path.isfile(file_path):
                try:
                    title, content = extract_values_from_file(file_path)
                    data.append([title, content, [0,1] if folder == "NonClickbait_Auto" else [1,0]])
                except (json.JSONDecodeError, KeyError) as e:
                    print(f"Error processing {filename}: {e}")
                    continue

# folder_path = './data/Part2'

# Print or use the lists as needed
print("data:", data[:5])
print(len(data))
```

```python
import csv

filename = "output.csv"

# Writing to CSV file
with open(filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    
    # Write each row
    writer.writerows(data)

print(f"Data has been written to {filename}")
```
